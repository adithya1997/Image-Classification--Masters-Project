```python
from keras.datasets import fashion_mnist
```

```python
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```python
print("Shape of x_train: {}".format(X_train.shape))
print("Shape of y_train: {}".format(y_train.shape))
```

```python
print("Shape of x_test: {}".format(X_test.shape))
print("Shape of y_test: {}".format(y_test.shape))
```

```python
#@title
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(10,12))
for i in range(100):
    plt.subplot(10,10,i+1)  
    plt.xticks([]) 
    plt.yticks([])
    plt.grid(False) 
    plt.imshow(X_train[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[y_train[i]])
```

```python
#@title
# Normalize the data
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255.0
X_test /= 255.0
```

```python
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1] * X_train.shape[2])
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1] * X_test.shape[2])
print(X_train.shape)
print(X_test.shape)
```

```python
# Decision Tree

from sklearn.tree import DecisionTreeClassifier
import time

beg = time.time()

#tree = DecisionTreeClassifier(max_depth=100, criterion='entropy')
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)

end = time.time()
dtree_time = end-beg

print("Time for Decision Tree: {:0.2f} minute".format(dtree_time/60.0))
```

```python
from sklearn import metrics
# Decision Tree report and analysis
y_pred_dtree = dtree.predict(X_test)
dtree_f1 = metrics.f1_score(y_test, y_pred_dtree, average= "weighted")
dtree_accu = metrics.accuracy_score(y_test, y_pred_dtree)
dtree_cm = metrics.confusion_matrix(y_test, y_pred_dtree)
print("-----------------Decision Tree Report---------------")
print("F1-Score:- {}".format(dtree_f1))
print("Accuracy Score:- {}".format(dtree_accu))
print("Confusion Matrix:- \n", dtree_cm)
```
