```python
from keras.datasets import fashion_mnist
```

```python
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```python
print("Shape of x_test: {}".format(X_test.shape))
print("Shape of y_test: {}".format(y_test.shape))
```

```python
#@title
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(10,12))
for i in range(100):
    plt.subplot(10,10,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(X_train[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[y_train[i]])
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255.0
X_test /= 255.0
```

```python
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1] * X_train.shape[2])
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1] * X_test.shape[2])
```

```python
from sklearn.ensemble import RandomForestClassifier
import time
# Random Forest
beg = time.time()

rforest = RandomForestClassifier(criterion='entropy', max_depth=70, n_estimators=100)
rforest.fit(X_train, y_train)

end = time.time()
rforest_time = end-beg
```

```python
# Random Forest report and analysis
from sklearn import metrics
y_pred_rforest = rforest.predict(X_test)
rforest_f1s = metrics.f1_score(y_test, y_pred_rforest, average= "weighted")
rforest_accu = metrics.accuracy_score(y_test, y_pred_rforest)
rforest_confmat = metrics.confusion_matrix(y_test, y_pred_rforest)
print("-----------------Random Forest Report---------------")
print("F1-Score:- {}".format(rforest_f1s))
print("Accuracy Score:- {}".format(rforest_accu))
print("Confusion Matrix:- \n", rforest_confmat)
```
